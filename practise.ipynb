```python
bios_new['heightcategory']=(bios_new['height_cm'].apply(lambda x: "Short" if x<175 else('average' if x<185 else 'Tall')))
```

```python
bios_new.head()
```

```python
def categorize_athlere(row):
    if row ['height_cm']<175 and row ['weight_kg']<70:
        return 'short'
    elif row ['height_cm']<185 and row ['weight_kg']<80:
        return 'average'
    else:
        return 'tall'

bios_new['height_weight_category']=bios_new.apply(categorize_athlere, axis=1)    
    
```
